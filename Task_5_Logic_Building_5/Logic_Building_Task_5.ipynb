 Problem 1: Social Media – Post Engagement Analyzer

 Problem Statement:

 Create a function to analyze engagement on a social media post.

 Rules:

 - Calculate total likes

 - If Total Likes ≥ 1000 → Viral Post

 - Otherwise → Normal Engagement

```python
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    # Check engagement status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    
    return total_likes, status

likes = [200, 300, 150, 400]
total, status = analyze_engagement(likes)
print("Total Likes:", total)
print("Post Status:", status)

```

 Problem 3: Agriculture – Rainfall Adequacy Checker

 Problem Statement:

 Analyze seasonal rainfall data.

 - Calculate average rainfall

 - Average ≥ required level → Adequate Rainfall

 - Otherwise → Inadequate Rainfall

```python
def rainfall_checker(rainfall_list, required_level):
    total = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total += rain
    
    average = total / len(rainfall_list)
    
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    
    return average, status

rain_data = [60, 75, 80, 70]
avg, rain_status = rainfall_checker(rain_data, 70)
print("Average Rainfall:", avg)
print("Rainfall Status:", rain_status)

```

 Problem 4: Social Media – Duplicate Account Detection

 Problem Statement:

 Detect duplicate usernames in a system.

 - If duplicates exist → Flag accounts

 - Else → Accounts are unique

```python
def detect_duplicates(usernames):
    # Convert list to set to remove duplicates
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        return "Yes"
    else:
        return "No"

users = ["john", "alice", "john", "bob"]
duplicate_status = detect_duplicates(users)
print("Duplicate Accounts Found:", duplicate_status)

```

 Problem 6: Agriculture – Premium Crop Price Filter

 Problem Statement:

 Filter crop prices above ₹2000.

 - Price > 2000 → Premium Crop

 - Else → Regular Crop

```python
def filter_premium_crops(prices):
    premium = []
    
    # Loop to filter prices
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    return premium

crop_prices = [1500, 2500, 1800, 3200]
premium_list = filter_premium_crops(crop_prices)
print("Premium Crops:", premium_list)

```

 Problem 7: System Monitoring – Application Health Checker

 Problem Statement:

 Evaluate application health based on error count.

 - Errors = 0 → Healthy

 - Errors ≤ 5 → Minor Issues

 - Errors > 5 → Critical Issues

 Problem 9: E-Learning – Student Attendance Eligibility System

 Problem Statement:

 Calculate attendance percentage and check exam eligibility.

 - Attendance ≥ 75% → Eligible

 - Attendance < 75% → Not Eligible

```python
def attendance_checker(attendance_list):
    present_count = 0
    
    # Count present days
    for day in attendance_list:
        if day == "P":
            present_count += 1
    
    percentage = (present_count / len(attendance_list)) * 100
    
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    return percentage, status

attendance_data = ["P", "P", "A", "P", "P"]
attendance_percent, exam_status = attendance_checker(attendance_data)
print("Attendance Percentage:", attendance_percent)
print("Exam Eligibility:", exam_status)

```

 Problem 10: Smart Electricity Bill Analyzer

 Problem Statement:

 Calculate electricity bill based on slab system.

 Billing Rules:

 - First 100 units → ₹3 per unit

 - Next 100 units → ₹5 per unit

 - Above 200 units → ₹7 per unit

 Classify usage based on total bill.

```python
def electricity_bill(units):
    bill = 0
    
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    
    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"
    
    return bill, usage

total_bill, usage_status = electricity_bill(250)
print("Total Bill:", total_bill)
print("Usage Status:", usage_status)
```
