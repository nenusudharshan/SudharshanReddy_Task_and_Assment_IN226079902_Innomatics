 Problem 1: Mobile Recharge Validation System



 Problem Statement:

 Create a function to validate a mobile recharge request.

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        try:
            amount = int(input("Enter recharge amount: "))
            
            # Check minimum recharge condition
            if amount < 50:
                print("Recharge amount must be at least ₹50.")
                continue
            
            # Check if plan is valid
            if amount in valid_plans:
                print("Recharge Successful")
                break
            else:
                print("Invalid plan. Valid plans are:", valid_plans)
                
        except ValueError:
            print("Please enter a valid number.")

# Call function
validate_recharge()

```

 Problem 2: Inventory Reorder Alert System

 Problem Statement:

 Given a dictionary of products and stock quantities:

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Sample inventory
inventory = {
    "Rice": 20,
    "Oil": 10,
    "Sugar": 8,
    "Salt": 25
}

check_inventory(inventory)
```

 Problem 3: Student Result Processing System

 Problem Statement:

 - Accept list of student marks

 - Calculate average

 - Print Pass if average ≥ 50 else Fail

```python
def process_result(marks):
    total = 0
    
    # Calculate total marks
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass"
    else:
        return "Fail"

# Example marks
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)
print("Result:", result)
```

 Problem 4: Cab Fare Estimator with Retry Option


 Problem Statement:

 - Base fare = ₹50

 - ₹12 per km

 - Peak hour → 25% extra

 - Allow retry using while loop

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)
    
    if peak.lower() == "yes":
        fare += fare * 0.25  # 25% extra
    
    return fare

while True:
    try:
        km = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")
        
        total_fare = calculate_fare(km, peak_hour)
        print("Total Fare: ₹", total_fare)
        
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break
            
    except ValueError:
        print("Please enter valid input.")

```

 Problem 5: Employee Attendance Eligibility Checker



 Problem Statement:

 - Accept attendance list ("P", "A")

 - Calculate attendance percentage

 - Return Eligible if ≥ 75% else Not Eligible

```python
def check_attendance(attendance_list):
    present_count = 0
    
    # Count present days
    for day in attendance_list:
        if day == "P":
            present_count += 1
    
    percentage = (present_count / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example attendance
attendance = ["P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance)
print("Status:", status)
```

 Problem 6: Password Strength Checker



 Problem Statement:

 Create a function to check password strength.



 Rules:

 - Minimum length = 8

 - Must contain at least one digit

 - Must contain at least one special character (@ # $)

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check minimum length
    if len(password) < 8:
        return "Weak Password  (Minimum length 8 required)"
    
    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password (Must include digit and special character)"

# Test password
user_password = input("Enter your password: ")
print(check_password(user_password))
```
