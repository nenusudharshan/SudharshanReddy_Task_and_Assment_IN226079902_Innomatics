# Problem 1: Unique Words in a Sentence
Problem Statement:

Given a sentence, identify all unique words and display:

The count of unique words

The unique words themselves


```python
# Taking input from user
sentence = input("Enter sentence: ")

# Splitting sentence into individual words
words = sentence.split()

# Converting list to set to remove duplicates
unique_words = set(words)

# Displaying results
print("\n--- Unique Words Analysis ---")
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)

```

# Problem 2: Highest Salary from Employee Data
Problem Statement:

Store employee names and their salaries,
determine which employee has the highest salary.


```python
# Dictionary storing employee salary data
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Initialize variables to track maximum salary
max_salary = 0
max_sal_name = ""

# Iterating through dictionary
for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        max_sal_name = name

# Display result
print("\n--- Salary Analysis ---")
print("Highest Salary:", max_sal_name, "-", max_salary)

```

# Problem 3: Find Maximum and Minimum Values
Problem Statement:

Find the maximum and minimum values from a list of numbers.


```python
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Display results
print("\n--- Number Analysis ---")
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)

```

# Problem 4: Count Products Above a Price Threshold
Problem Statement:

Count how many products have a price greater than 1000.


```python
prices = [450, 1200, 899, 1500, 300]

count = 0

# Loop through price list
for price in prices:
    if price > 1000:
        count += 1

print("\n--- Product Price Analysis ---")
print("Products above 1000:", count)

```

# Problem 5: Calculate Attendance Percentage
Problem Statement:

Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).


```python
attendance = ["P", "P", "A", "P", "P"]

present_count = 0

# Counting present days
for day in attendance:
    if day == "P":
        present_count += 1

# Calculating percentage
attendance_percentage = (present_count / len(attendance)) * 100

print("\n--- Attendance Analysis ---")
print("Attendance Percentage:", attendance_percentage)

```

# Problem 7: Count Character Frequency
Problem Statement:

Count how many times each character appears in a given string.


```python
text = "python"

char_frequency = {}

# Counting character occurrences
for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

print("\n--- Character Frequency Analysis ---")
print(char_frequency)

```

# Problem 10: Calculate Average Marks
Problem Statement:

Calculate the average marks of a student from a list of marks.


```python
marks = [70, 75, 80, 65]

total = 0

# Summing marks
for mark in marks:
    total += mark

# Calculating average
average = total / len(marks)

print("\n--- Marks Analysis ---")
print("Average Marks:", average)
```
